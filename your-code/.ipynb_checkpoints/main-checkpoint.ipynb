#### Retrieve an arbitary Wikipedia page of "Python" and create a list of links on that page

````python
# This is the url you will scrape in this exercise
#url ='https://en.wikipedia.org/wiki/Python' 
url = 'https://en.wikipedia.org/wiki/Python_(programming_language)'
html = requests.get(url).content
soup = BeautifulSoup(html)
#soup
````

````python
link = soup.find_all('a')
links= []
for l in link:
    get_url = str(l.get('href'))
    if get_url.startswith('/'):
        full_url = 'https://es.wikipedia.org' + get_url
        links.append(full_url)
links
    
````

#### Count number of tweets by a given Twitter account.

You will need to include a ***try/except block*** for account names not found. 
<br>***Hint:*** the program should count the number of tweets for any provided account

````python
# This is the url you will scrape in this exercise 
# You will need to add the account credentials to this url
url = 'https://twitter.com/Dany_Kino'
html = requests.get(url).content
soup = BeautifulSoup(html)
````

````python
#your code
try:
    tweets = soup.find('li',{'class':'ProfileNav-item ProfileNav-item--tweets is-active'})
    number_of_tweets= tweets.find('a').find('span',{'class':'ProfileNav-value'})
    #print(number_of_tweets)
    print('Number of tweets made by Dany Kino: ', number_of_tweets.get('data-count'))
except:
    print('account names not found')
````

#### List all language names and number of related articles in the order they appear in wikipedia.org

````python
# This is the url you will scrape in this exercise
url = 'https://www.wikipedia.org/'
html = requests.get(url).content
soup = BeautifulSoup(html)
````

````python
#your code
languages = soup.find_all('a',{'class','link-box'})
#languages
articles_per_language = []
for language in languages:
    articles = str(language.text.replace('\n',' ').replace('+',''))
    #articles_per_language.append(str(language.text.replace('\n',' ').replace('+','')))
    #articles_per_language.append(language.translate(str.maketrans({'\n': ' ', '+': '', '\xa0000':''})))
    #print(articles)
    articles_per_language.append(articles)
#articles_per_language
colname = ['Articles per Language']
df = pd.DataFrame(articles_per_language, columns = colname)
df
````

#### Top 10 languages by number of native speakers stored in a Pandas Dataframe

````python
# This is the url you will scrape in this exercise
url = 'https://en.wikipedia.org/wiki/List_of_languages_by_number_of_native_speakers'
html = requests.get(url).content
soup = BeautifulSoup(html)
````

````python
#your code
languages_table = soup.find_all('table', {'class':'sortable'})
languages_table = languages_table[0]
#table
row_languages = languages_table.find_all('tr')
row_languages = [language.text.replace('\n\n', '\n').strip().split('\n') for language in row_languages]
#row_languages[0]
colnames = row_languages[0]
del colnames[4:6]
data = row_languages[1:11]
df = pd.DataFrame(data, columns = colnames)
df.head(10)
````
